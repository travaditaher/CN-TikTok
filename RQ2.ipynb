### RQ2: English Language

```python
import pandas as pd
import re

english_category_counts = {
    'inclined towards peace': 0,
    'unrelated/cannot infer sentiment towards war': 0,
    'inclined towards ukraine': 0,
    'inclined towards russia': 0,
    'neutral sentiment': 0
}
total_english_comments = 0

# Load the data from CSV file
df = pd.read_csv('/Users/tahers/Documents/CN-TikTok/Test.csv')  

# Get first 10 comments from the 'Comments' column
comments = df['Comments'].tolist()
other_comments = df['Other Languages than English'].tolist()
# print(other_comments)
english_sentiments = df['Sentiments'].tolist()
i=0
# Iterate over each comment
for comment_str, other_comment_str, sentiment_str in zip(comments, other_comments,english_sentiments):
    i+=1
    print(i)
    if pd.isna(other_comment_str) or other_comment_str.strip() == '':
        continue
    # Split the comment string based on numbering pattern using regular expression
    individual_comments = re.split(r'\b\d+\.\s*', comment_str)
    individual_comments = [comment.strip() for comment in individual_comments if comment.strip()]

    individual_other_comments = re.split(r'\b\d+\.\s*', other_comment_str)
    individual_other_comments = [comment.strip() for comment in individual_other_comments if comment.strip()]

    total_english_comments += len(individual_comments)

    english_sentiments = re.split(r'\b\d+\)\s*', sentiment_str)
    english_sentiments = [comment.strip().lower() for comment in english_sentiments if comment.strip()]

    # Compare each individual comment with comments in 'Other Languages than English' column
    for comment, sentiment in zip(individual_comments, english_sentiments):
        if comment not in individual_other_comments:
            # Increment corresponding sentiment count in the dictionary
            if sentiment in english_category_counts:
                english_category_counts[sentiment] += 1
            else:
                print(f"Unknown sentiment: {sentiment}")     

# Calculate percentage of each sentiment category
english_category_percentages = {
    sentiment: count / total_english_comments * 100 if total_english_comments > 0 else 0
    for sentiment, count in english_category_counts.items()
}

# Print the sentiment counts and percentages
print("English Sentiment Counts:")
for sentiment, count in english_category_counts.items():
    print(f"{sentiment}: {count}")

print("\nEnglish Sentiment Percentages:")
for sentiment, percentage in english_category_percentages.items():
    print(f"{sentiment}: {percentage:.2f}%")
```

### RQ2: Other Language

```python
import pandas as pd
import re

def calculate_category_percentages(input_file):
    # Initialize counters for each category
    category_counts = {
        'inclined towards ukraine': 0,
        'inclined towards russia': 0,
        'neutral sentiment': 0,
        'inclined towards peace': 0,
        'unrelated/cannot infer sentiment towards war': 0,
    }
    
    # Read CSV file into DataFrame
    df = pd.read_csv(input_file)
    
    # Iterate over each row
    for index, row in df.iterrows():
        # Extract sentiments from the 'Comments_sentiments' column
        sentiments = str(row['Sentiments'])
        
        # Find all sentiments with numbering removed
        clean_sentiments = re.findall(r'\d+[\.\)]\s*([^\d]+)', sentiments)
        
        # Iterate over each sentiment
        for sentiment in clean_sentiments:
            # Split the sentiment by comma (if present) to handle multiple sentiments
            individual_sentiments = [s.strip() for s in sentiment.split(',')]
            for ind_sentiment in individual_sentiments:
                ind_sentiment = ind_sentiment.lower()
                # Increment the counter for the corresponding category
                if ind_sentiment == 'cannot infer sentiment towards war':
                    ind_sentiment = 'unrelated/cannot infer sentiment towards war'
                if ind_sentiment == 'neutral':
                    ind_sentiment = 'neutral sentiment'
                if ind_sentiment == 'inclined towards ukraine.' or ind_sentiment =='inclined towards ukarine':
                    ind_sentiment = 'inclined towards ukraine'
                if ind_sentiment == 'inclined towards georgia and ukraine':
                    ind_sentiment = 'inclined towards ukraine'
                if ind_sentiment == 'inclinded towards peace':
                    ind_sentiment = 'inclined towards peace'

                category_counts[ind_sentiment] += 1
    
    # Calculate total number of comments
    total_comments = sum(category_counts.values())
    print(total_comments)
    print(category_counts)
    
    # Calculate percentage for each category and filter out categories with zero counts
    category_percentages = {}
    for category, count in category_counts.items():
        if count > 0:
            percentage = (count / total_comments) * 100
            category_percentages[category] = percentage
    
    return category_percentages

# Provide the path to your input CSV file
input_file = '/Users/tahers/Documents/CN-TikTok/RQ3_Data.csv'

# Calculate category percentages
percentages = calculate_category_percentages(input_file)

# Print the category percentages and categories with non-zero counts
print("Category Percentages:")
for category, percentage in percentages.items():
    print(f"{category}: {percentage:.2f}%")
```
